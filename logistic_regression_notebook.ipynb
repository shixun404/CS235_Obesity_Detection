#### Feature Engineering

###### 1、Single dimensional feature selection (based on variance, to see if a feature has a certain value that is over 90%)

```python
def func1(df, ratio = 0.9):
    dropList = []
    dfLen = len(df)
    for column in df.columns:
        columnmaxnums = df[column].value_counts().iloc[0]
        columnmaxnumsratio = columnmaxnums / dfLen
        print('{} features, maximum value proportion: {}'.format(column, round(columnmaxnumsratio,3)))
        if columnmaxnumsratio > ratio:
            dropList.append(column)
    return dropList
```

```python
def func2(df):
    noOneHotList = []
    for column in df.columns:
        columnNumsLen = len(df[column].value_counts())
        print('{}features, the number of different data is: {}'.format(column, columnNumsLen))
        if columnNumsLen == 2:
              noOneHotList.append(column)
    return noOneHotList
```

###### 5、Data Degradation PCA

```python
class PCA(object):
    def __init__(self,n=5):
        self.n = n
        self.meanArray = None
        self.vArray = None
    def fit(self,data):
        self.meanArray = np.mean(data, axis=0)
        xArray = np.subtract(data, self.meanArray)
        covArray = np.dot(xArray.T, xArray)
        wArray, vArray = np.linalg.eig(covArray)
        wArray = wArray.real
        self.vArray = vArray.real
    def transform(self,data):
        arraylist = []
        xArray = np.subtract(data, self.meanArray)
        for i in range(self.n):
            arraylist.append(np.dot(xArray, self.vArray.T[i]))
        return np.array(arraylist).T
```

###### 6、Data normalization

```python
class MinMaxScaler(object):
    def __init__(self,feature_range=[0,1]):
        self.feature_range = feature_range
        self.dataMin = None
        self.dataMax = None
    def fit(self,data):
        self.dataMin = data.min(axis=0)
        self.dataMax = data.max(axis=0)
    def transform(self,data):
        x_std = (data - self.dataMin)/(self.dataMax - self.dataMin)
        x_scaled = x_std * (self.feature_range[1] - self.feature_range[0]) + self.feature_range[0]
        return x_scaled
```
